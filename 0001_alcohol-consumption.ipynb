## Hypothesis: Most Popular Country Per Drinks

```python
mean = consumption['beer_servings'].mean()
print(mean)
consumption.sort_values(by='beer_servings', ascending=False)
```

```python

for col in consumption.columns[1:]: 
    drink = col
    sorted_consumption = consumption.sort_values(by=drink, ascending=False)
    g = sns.barplot(data=sorted_consumption.head(n=10), orient='h', x=drink, y='country', color='#30a2da')
    plt.axvline(consumption[drink].mean(), color='#fc4f30',linestyle='dashed', linewidth=2)
    g.set_title("Top 10 Country Serving " + drink + " \n (red is average)")
    plt.show()
```

```python

for col in consumption.columns[1:]: 
    drink = col
    sorted_consumption = consumption.sort_values(by=drink, ascending=False)
    plt.xlim(0, 450)
    g = sns.barplot(data=sorted_consumption.head(n=10), orient='h', x=drink, y='country', color='#30a2da')
    plt.axvline(consumption[drink].mean(), color='#fc4f30',linestyle='dashed', linewidth=2)
    g.set_title("Top 10 Country Serving " + drink + " \n (red is average)")
    plt.show()
```

```python

sorted_consumption = consumption.sort_values(by='beer_servings', ascending=False)
g = sns.barplot(data=sorted_consumption.tail(n=10), orient='h', x='beer_servings', y='country', color='#30a2da')
plt.axvline(mean, color='#fc4f30',linestyle='dashed', linewidth=2)
g.set_title("Bottom 10 Country Serving Beer \n (red is average)")
```

```python
for col in consumption.columns[1:]: 
    drink = col
    sorted_consumption = consumption.sort_values(by=drink, ascending=False)
    plt.xlim(0, 450)
    g = sns.barplot(data=sorted_consumption.tail(n=10), orient='h', x=drink, y='country', color='#30a2da')
    plt.axvline(consumption[drink].mean(), color='#fc4f30',linestyle='dashed', linewidth=2)
    g.set_title("Bottom 10 Country Serving " + drink + " \n (red is average)")
    plt.show()
```
