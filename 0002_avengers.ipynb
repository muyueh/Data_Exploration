## `countif`

```python

avengers['Deaths'] = avengers[['Death1', 'Death2', 'Death3', 'Death4', 'Death5']].apply((lambda list: sum([1 if x == "YES" else 0 for x in list])), axis=1)

avengers['Death-Year'] = avengers['Deaths'] / avengers['Years since joining']
avengers['Death-Appearance'] = avengers['Deaths'] / avengers['Appearances']
avengers[['Death-Year', 'Death-Appearance', 'Deaths','Years since joining','Appearances']].head()
```

```python
# avengers['Years since joining'].hist(weights=)

avengers['yearbins'] = (avengers['Years since joining'] / 10).round() * 10
g = sns.barplot(x=avengers['yearbins'], y=avengers['Deaths'], ci=None)

for p in g.patches:
    g.annotate(s="{0:.1f}%".format(p.get_height() / 1 * 100 ) , xy=(p.get_x() + (p.get_width() / 2), p.get_height() * 1.005), ha='center')    

g.set_title(">80% Death Rate for Avengers in 1970 and 1950 ")
g.set(ylabel="Death Times / Number")


```

```python
def Chart2Number (x): 
    return 1 if x == 'YES' else 0
g = sns.barplot(x=avengers['Gender'], y=avengers['Death1'].map(Chart2Number) , ci=None)

for p in g.patches:
    g.annotate(s="{0:.1f}%".format(p.get_height() / 1 * 100 ) , xy=(p.get_x() + (p.get_width() / 2), p.get_height() * 1.005), ha='center')    

g.set_title("Slightly Higher Death Rate for Male")
g.set(ylabel="Death / Number")

```

```python
g = sns.barplot(x=avengers['Honorary'], y=avengers['Death1'].map(Chart2Number), ci=None )

for p in g.patches:
    g.annotate(s="{0:.1f}%".format(p.get_height() / 1 * 100 ) , xy=(p.get_x() + (p.get_width() / 2), p.get_height() * 1.005), ha='center')    

g.set_title(">40% Mortality for Full Time Avengers")
g.set(ylabel="Death / Number")



```

## Return Rate

```python

died = avengers[avengers['Death1'] == 'YES']

def Chart2Number (x): 
    return 1 if x == 'YES' else 0
g = sns.barplot(x=died['Gender'], y=died['Return1'].map(Chart2Number) , ci=None)
for p in g.patches:
    g.annotate(s="{0:.1f}%".format(p.get_height() / 1 * 100 ) , xy=(p.get_x() + (p.get_width() / 2), p.get_height() * 1.005), ha='center')    

g.set_title("Once died, female has slightly more change to return")
g.set(ylabel="Return / Death")


```

```python
g = sns.barplot(x=died['Honorary'], y=died['Return1'].map(Chart2Number) , ci=None)
g.set_title("Academy Avengers never return; probationary always does")
g.set(ylabel="Return / Death")

for p in g.patches:
    g.annotate(s="{0:.1f}%".format(p.get_height() / 1 * 100 ) , xy=(p.get_x() + (p.get_width() / 2), p.get_height() * 1.005), ha='center')    

```
